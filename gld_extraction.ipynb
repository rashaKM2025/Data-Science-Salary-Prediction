```python
df = pd.read_csv('/content/drive/MyDrive/glassdoor/glassdoorjobs_normalization.csv')
df
```

```python
def extract_seniority(title):
    title = str(title).lower()
    if "senior" in title or "sr" in title:
        return "senior"
    elif "lead" in title:
        return "lead"
    elif "manager" in title:
        return "manager"
    elif "junior" in title or "jr" in title:
        return "junior"
    else:
        return "mid"

df['job_seniority'] = df['job_title'].apply(extract_seniority)
```

```python
west_coast = ['ca','wa','or']
east_coast = ['ny','nj','ma','md','va','dc','pa','nc','sc','fl']
midwest = ['il','oh','mi','mn','wi','in','mo','ia','ks','ne']
south = ['tx','ok','la','ga','al','tn','ky','ms','ar']

def map_region(state):
    state = str(state).lower()
    if state in west_coast:
        return "west_coast"
    elif state in east_coast:
        return "east_coast"
    elif state in midwest:
        return "midwest"
    elif state in south:
        return "south"
    else:
        return "other"

df['region'] = df['state'].apply(map_region)
```

```python
tech_hubs = ['san francisco','seattle','new york','mountain view','palo alto','menlo park','redmond']
df['is_tech_hub'] = df['city'].apply(lambda x: str(x).lower() in tech_hubs)
```

```python
df['rating_norm'] = df['rating'] / 5.0
df['is_top_rated'] = df['rating'].apply(lambda x: 1 if x >= 4.0 else 0)
```

```python
df.to_csv('/content/glassdoorjobs_extraction.csv', index=False)
```
