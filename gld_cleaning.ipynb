```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/glassdoor/glassdoor_jobs.csv', encoding='latin1')
df.head()
```

```python
df = df.dropna(thresh=len(df.columns)//2)
```

```python
df = df.fillna({'Salary Estimate': 'Unknown', 'Company Name': 'Unknown', 'Job Title': 'Unknown'})
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ','_')
```

```python
df = df.drop_duplicates()
```

```python
df = df.drop_duplicates(subset=['job_title', 'company_name'])
```

```python
df.shape
```

```python
def parse_salary(s):
  if pd.isna(s) or s == 'Unknown':
    return np.nan, np.nan
  s = s.replace('$','').replace('K','').replace(',','').replace('per hour','').replace('Employer Provided Salary:','').strip()
  parts = s.split('-')
  try:
    if len(parts) == 2:
      return int(parts[0]), int(parts[1])
    else:
      return int(parts[0]), int(parts[0])
  except:
    return np.nan, np.nan

df[['min_salary', 'max_salary']] = df['salary_estimate'].apply(lambda x: pd.Series(parse_salary(str(x))))
```

```python
df['company_name_clean'] = df['company_name'].str.extract(r'([A-Za-z\s&]+)')
df['rating'] = df['company_name'].str.extract(r'(\d+\.\d+)')
```

```python
if 'job_type' in df.columns:
    df['job_type_encoded'] = df['job_type'].astype('category').cat.codes
```

```python
df['avg_salary'] = df[['min_salary','max_salary']].mean(axis=1)
```

```python
df.to_csv('/content/glassdoorjobs_cleaning.csv', index=False)
```
